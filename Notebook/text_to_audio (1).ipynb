```python
import gradio as gr
from gtts import gTTS
import speech_recognition as sr
from pydub import AudioSegment
import os

# -------- SPEECH TO TEXT FUNCTION --------
def speech_to_text(audio):
    recognizer = sr.Recognizer()

    audio = AudioSegment.from_file(audio)
    audio.export("temp.wav", format="wav")

    with sr.AudioFile("temp.wav") as source:
        audio_data = recognizer.record(source)

    try:
        text = recognizer.recognize_google(audio_data)
        return text
    except:
        return "Sorry, could not recognize speech."


# -------- TEXT TO SPEECH FUNCTION --------
def text_to_speech(text, language):
    tts = gTTS(text=text, lang=language)
    tts.save("output.mp3")
    return "output.mp3"


# -------- COMBINED FUNCTION --------
def voice_ai(audio, text, language):
    recognized_text = ""

    if audio is not None:
        recognized_text = speech_to_text(audio)
        text = recognized_text

    audio_file = text_to_speech(text, language)
    return recognized_text, audio_file


# -------- GRADIO UI --------
ui = gr.Interface(
    fn=voice_ai,
    inputs=[
        gr.Audio(type="filepath", label="üéôÔ∏è Upload Speech (Speech-to-Text)"),
        gr.Textbox(label="‚úçÔ∏è Enter Text (Text-to-Speech)"),
        gr.Dropdown(
            choices=[
                ("English", "en"),
                ("Tamil", "ta"),
                ("Hindi", "hi"),
                ("French", "fr"),
                ("German", "de"),
                ("Spanish", "es")
            ],
            label="üåç Select Language",
            value="en"
        )
    ],
    outputs=[
        gr.Textbox(label="üìù Recognized Text"),
        gr.Audio(label="üîä Generated Voice")
    ],
    title="üé§ Advanced Voice AI System",
    description="Speech-to-Text + Multi-Language Voice Generation using AI"
)

ui.launch()
```

```python
import gradio as gr
import whisper
from gtts import gTTS
from deep_translator import GoogleTranslator
import os

# Load Whisper model
model = whisper.load_model("base")

# ---------- SPEECH TO TEXT ----------
def speech_to_text(audio_path):
    result = model.transcribe(audio_path)
    return result["text"]

# ---------- TRANSLATION ----------
def translate_text(text, target_lang):
    if target_lang == "en":
        return text
    return GoogleTranslator(source='auto', target=target_lang).translate(text)

# ---------- EMOTION VOICE ----------
def emotion_voice(text, emotion, lang):
    slow = False
    if emotion == "Sad":
        slow = True

    tts = gTTS(text=text, lang=lang, slow=slow)
    tts.save("output.mp3")
    return "output.mp3"

# ---------- MAIN FUNCTION ----------
def voice_ai(audio, text, target_lang, emotion):
    recognized_text = ""

    if audio:
        recognized_text = speech_to_text(audio)
        text = recognized_text

    translated_text = translate_text(text, target_lang)
    audio_out = emotion_voice(translated_text, emotion, target_lang)

    return recognized_text, translated_text, audio_out

# ---------- UI ----------
app = gr.Interface(
    fn=voice_ai,
    inputs=[
        gr.Audio(type="filepath", label="üéôÔ∏è Speak Now"),
        gr.Textbox(label="‚úçÔ∏è Or Enter Text"),
        gr.Dropdown(
            choices=["en", "ta", "hi", "fr", "de", "es"],
            value="en",
            label="üåç Output Language"
        ),
        gr.Radio(["Neutral", "Happy", "Sad"], value="Neutral", label="üòä Emotion")
    ],
    outputs=[
        gr.Textbox(label="üìù Recognized Text"),
        gr.Textbox(label="üåê Translated Text"),
        gr.Audio(label="üîä Voice Output")
    ],
    title="üî• Advanced Voice AI System",
)

app.launch()
```

```python
import gradio as gr
import whisper
from gtts import gTTS
from deep_translator import GoogleTranslator
import os

# Load Whisper model
model = whisper.load_model("base")

# -------- Speech to Text (Auto Language Detect) --------
def speech_to_text(audio_path):
    result = model.transcribe(audio_path)
    return result["text"], result["language"]

# -------- Translation --------
def translate_text(text, target_lang):
    if target_lang == "auto":
        return text
    return GoogleTranslator(source='auto', target=target_lang).translate(text)

# -------- Emotion Voice --------
def generate_voice(text, lang, emotion):
    slow = True if emotion == "Sad" else False
    tts = gTTS(text=text, lang=lang, slow=slow)
    tts.save("output.mp3")
    return "output.mp3"

# -------- Main Function --------
def voice_ai(audio, text, target_lang, emotion):
    detected_lang = ""

    if audio:
        text, detected_lang = speech_to_text(audio)

    translated = translate_text(text, target_lang)
    voice = generate_voice(translated, target_lang if target_lang != "auto" else detected_lang, emotion)

    return text, detected_lang, translated, voice

# -------- UI --------
app = gr.Interface(
    fn=voice_ai,
    inputs=[
        gr.Audio(type="filepath", label="üéôÔ∏è Speak in ANY Language"),
        gr.Textbox(label="‚úçÔ∏è Or Type Text"),
        gr.Dropdown(
            choices=[
                "auto","en","ta","hi","te","ml","kn","fr","de","es",
                "zh-cn","ja","ko","ru","ar","pt","it"
            ],
            value="auto",
            label="üåç Output Language"
        ),
        gr.Radio(["Neutral","Happy","Sad"], value="Neutral", label="üòä Emotion")
    ],
    outputs=[
        gr.Textbox(label="üìù Recognized Text"),
        gr.Textbox(label="üåê Detected Language"),
        gr.Textbox(label="üîÅ Translated Text"),
        gr.Audio(label="üîä Voice Output")
    ],
    title="üåç Universal Voice AI (All Languages)",
)

app.launch()
```

```python
import librosa
import numpy as np

def detect_emotion(audio_path):
    y, sr = librosa.load(audio_path)

    pitch = np.mean(librosa.yin(y, fmin=50, fmax=300))
    energy = np.mean(librosa.feature.rms(y=y))
    zcr = np.mean(librosa.feature.zero_crossing_rate(y))

    if energy > 0.05 and pitch > 180:
        return "Happy"
    elif energy < 0.02:
        return "Sad"
    else:
        return "Neutral"
```
